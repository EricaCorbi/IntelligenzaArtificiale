# Challenge Telco Communication

Ho a disposizione un dataset con le caratteristiche dei clienti, voglio costruire un classificatore che prevede se un cliente resterà tale o meno. La fetures target nel dataset è il Churn,che contiene Yes o No in base alle caratteristiche del cliente che sono tutte le altre features del dataset.

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split #per dividere il dataset in dati di training e dati di test 
from sklearn.tree import DecisionTreeClassifier #per usare un classificatore Decision Tree
from sklearn.metrics import accuracy_score  #per usare come metrica l'accuratezza
from sklearn.metrics import confusion_matrix #per usare una matrice di confusione per misurare le prestazioni 
from sklearn.preprocessing import LabelEncoder #per trasformare i tipi in int
from sklearn.metrics import classification_report
```

```python
dataset = pd.read_csv('../Challenge2/WA_Fn-UseC_-Telco-Customer-Churn.train.csv') #importo i dati 
```

Importo il dataset che si trova nella cartella Challenge2 

```python
dataset.head()
```

Visualizzo le prime 5 righe del dataset

```python
dataset.shape #ci sono 5634 righe e 21 colonne
```

Utilizzo questo metodo per controllare se ci sono dei valori nulli nelle righe del dataset 

```python
dataset.isna().any() #non ci sono righe nulle 
```

```python
features_categoriche = dataset.select_dtypes(include = 'object') #selectiono le features che sono object
features_categoriche
```

```python
features_numeriche = dataset.select_dtypes(include = np.number)   #seleziono le features che sono numeri 
features_numeriche 
```

```python
# utilizzo il label encoder per trasformare i valori che non sono numerici in valori numerici
label_encoder = LabelEncoder()

for i in features_categoriche:
    features_categoriche[i] = label_encoder.fit_transform(features_categoriche[i])
    
features_categoriche
```

Il labelEncoder trasforma i valori di tipo object in valori interi

```python
newDataset = pd.concat([features_numeriche, features_categoriche], axis = 1) #concateno di nuovo le fatures numeriche con quelle trasformate
```

```python
newDataset #mostro il nuovo dataset
```

```python
#dichiaro quale feature è X e quale è y
X = newDataset.drop(columns=['Churn']) #X sono tutte le colonne, quindi tutte le feature del cliente 
y = newDataset.Churn #y è la colonna target Churn
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=0)
```

Divido il train e il test set, con l'insieme di test al 20%

```python
#decido di usare un classificatore DecisionTree
tree = DecisionTreeClassifier(criterion='entropy',random_state=1,max_depth=6)
tree.fit(X_train,y_train)
predictions = tree.predict(X_test)
```

```python
print("Accuracy : "+str(accuracy_score(y_test,predictions)))
```

Con il DecisionTree raggiungo un'accuratezza del 78,8%

```python
confusion_matrix = confusion_matrix(y_test, predictions)
print(confusion_matrix)
```

La matrice di confusione mostra 764 e 125 sono previsioni giuste e 78 e 160 sono previsioni errate

```python
print(classification_report(y_test,predictions))
```

Provo ora ad utilizzare una random forest per vedere se aumenta l'accuratezza

### La foresta casuale (o Random Forest) è un tipo di algoritmo di apprendimento automatico supervisionato basato sull’apprendimento di più modelli previsionali per formare un unico modello di previsione più potente.Ogni modello utilizzato dalla previsione del Random Forest di solito è un albero decisionale.
In questo caso decido di utilizzare 100 alberi decisionali

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_classifier = RandomForestClassifier(criterion='entropy',n_estimators=100, random_state=1)
random_classifier.fit(X_train,y_train)
y_pred = random_classifier.predict(X_test)
```

```python
print("Accuracy : "+str(accuracy_score(y_test,y_pred)))
```

L'accuratezza sale all'80%

```python
print(classification_report(y_test,y_pred))
```
